```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer

%matplotlib inline

sns.set_style("whitegrid")

plt.rcParams['font.family'] = 'DejaVu Sans'
```

```python
cancer = load_breast_cancer()
df = pd.DataFrame(cancer.data, columns=cancer.feature_names)
df['target'] = cancer.target

df['diagnosis'] = df['target'].map({0: 'Malign', 1: 'Benign'})

print("Veri setinin boyutu:", df.shape)
df.head()
```

```python
print("\nTanımlayıcı İstatistikler:")
df.describe()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='diagnosis')
plt.title('Tanı Dağılımı')
plt.xlabel('Tanı')
plt.ylabel('Hasta Sayısı')
plt.show()

print("\nTanı Dağılımı (%):\n")
print(df['diagnosis'].value_counts(normalize=True).mul(100).round(2))
```

```python
numeric_df = df.drop(['target', 'diagnosis'], axis=1)

plt.figure(figsize=(15, 10))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Özellikler Arası Korelasyon Matrisi')
plt.show()
```

```python
important_features = ['mean radius', 'mean texture', 'mean perimeter', 'mean area']

fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Önemli Özelliklerin Tanıya Göre Dağılımı')

for i, feature in enumerate(important_features):
    row = i // 2
    col = i % 2
    sns.boxplot(data=df, x='diagnosis', y=feature, ax=axes[row, col])
    axes[row, col].set_title(feature)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 8))
diagnosis_counts = df['diagnosis'].value_counts()
colors = ['#FF9999', '#66B2FF']
plt.pie(diagnosis_counts, labels=diagnosis_counts.index, autopct='%1.1f%%', 
        colors=colors, startangle=90)
plt.title('Meme Kanseri Tanı Dağılımı')
plt.axis('equal')
plt.show()
```

```python
# Hipotez Testleri
from scipy import stats

print("HIPOTEZ TESTLERİ\n")

# Hipotez 1: Malign ve Benign tümörlerin ortalama yarıçapları arasında anlamlı bir fark var mı?
print("Hipotez 1: Malign ve Benign tümörlerin ortalama yarıçapları arasında fark")
malign_radius = df[df['diagnosis'] == 'Malign']['mean radius']
benign_radius = df[df['diagnosis'] == 'Benign']['mean radius']

t_stat, p_value = stats.ttest_ind(malign_radius, benign_radius)
print(f"p-değeri: {p_value:.10f}")
print(f"Sonuç: {'İstatistiksel olarak anlamlı bir fark VAR' if p_value < 0.05 else 'İstatistiksel olarak anlamlı bir fark YOK'}")
print("\n" + "-"*50 + "\n")

# Hipotez 2: Malign ve Benign tümörlerin ortalama alanları arasında anlamlı bir fark var mı?
print("Hipotez 2: Malign ve Benign tümörlerin ortalama alanları arasında fark")
malign_area = df[df['diagnosis'] == 'Malign']['mean area']
benign_area = df[df['diagnosis'] == 'Benign']['mean area']

t_stat, p_value = stats.ttest_ind(malign_area, benign_area)
print(f"p-değeri: {p_value:.10f}")
print(f"Sonuç: {'İstatistiksel olarak anlamlı bir fark VAR' if p_value < 0.05 else 'İstatistiksel olarak anlamlı bir fark YOK'}")
print("\n" + "-"*50 + "\n")

# Hipotez 3: Tümör boyutu ile doku özellikleri arasında anlamlı bir korelasyon var mı?
print("Hipotez 3: Tümör boyutu ile doku özellikleri (mean texture) arasında korelasyon")
correlation, p_value = stats.pearsonr(df['mean radius'], df['mean texture'])
print(f"Korelasyon katsayısı: {correlation:.3f}")
print(f"p-değeri: {p_value:.10f}")
print(f"Sonuç: {'İstatistiksel olarak anlamlı bir korelasyon VAR' if p_value < 0.05 else 'İstatistiksel olarak anlamlı bir korelasyon YOK'}")
```

```python
# Hipotez sonuçlarını görselleştirme
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Yarıçap karşılaştırması
sns.boxplot(data=df, x='diagnosis', y='mean radius', ax=axes[0])
axes[0].set_title('Tanıya Göre Ortalama Yarıçap Dağılımı')
axes[0].set_xlabel('Tanı')
axes[0].set_ylabel('Ortalama Yarıçap')

# Alan karşılaştırması
sns.boxplot(data=df, x='diagnosis', y='mean area', ax=axes[1])
axes[1].set_title('Tanıya Göre Ortalama Alan Dağılımı')
axes[1].set_xlabel('Tanı')
axes[1].set_ylabel('Ortalama Alan')

plt.tight_layout()
plt.show()

# Korelasyon görselleştirmesi
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='mean radius', y='mean texture', hue='diagnosis')
plt.title('Tümör Boyutu ve Doku Özellikleri İlişkisi')
plt.xlabel('Ortalama Yarıçap (mean radius)')
plt.ylabel('Ortalama Doku (mean texture)')
plt.show()
```
